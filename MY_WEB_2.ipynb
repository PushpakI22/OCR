```python
# there are few steps to extract data from webpages first one is (import request) its allowes us to scrap  

import requests
url = "https://www.summet.com/dmsi/html/codesamples/addresses.html"
webpage = requests.get(url)
webpage.text
```

```python
url = "https://www.summet.com/dmsi/html/codesamples/addresses.html"
webpage = requests.get(url)
data = webpage.text
contact_no = re.findall(r'\(\d{3}\)\s\d{3}[-]\d{4}',data)
pin_no = re.findall(r'\b\d{5}\b',data)
names = re.findall(r'<li>([a-zA-Z]+\s[a-zA-Z]+)<br/>',data)
len(contact_no),len(pin_no),len(names)
print(contact_no)
print(pin_no)
print(names)
```

```python
#create a standard form by using pandas 
```

```python
dct = {"Names": names,
       "Contacts": contact_no,
       "Pin":pin_no}
df=pd.DataFrame(dct)
df
```

```python
#export to e-xcel 
dct = {"Names": names,
       "Contacts": contact_no,
       "Pin":pin_no}
df=pd.DataFrame(dct)
df.to_excel("webscraping_2 data.xlsx")  
```
